```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import dbscan_lab_helper as helper
```

```python
df = pd.read_csv('Blobs.csv')
data_1 = df.values
```

```python
helper.plot_dataset(data_1)
```

```python
from sklearn import cluster

dbscan = cluster.DBSCAN()
cluster_1 = dbscan.fit_predict(data_1)
```

```python
helper.plot_clustered_dataset(data_1 , cluster_1) 
```

```python
epsilion = 2
dbscan = cluster.DBSCAN(eps=epsilion)
cluster_2 = dbscan.fit_predict(data_1)

helper.plot_clustered_dataset(data_1 , cluster_2 , neighborhood=True) 
```
