# Why does SSE decrease over time?

Because Gradient Descent updates the model parameters in a way that reduces the prediction error at each iteration, so the squared errors become smaller.

# What is convergence?

Convergence occurs when parameter updates become negligible and the loss no longer decreases significantly.

# Experimentation 
1. Train the model with: 
a very large learning rate 


```python
model_large_lr = LinearRegression_Hazem(alpha=0.1, num_iterations=100)
model_large_lr.fit(x, y)
model_large_lr.plot(x,y)
```

1. Train the model with: 
a very small learning rate

```python
model_small_lr = LinearRegression_Hazem(alpha=0.00001, num_iterations=200)
model_small_lr.fit(x, y)
model_small_lr.plot(x,y)
```

 * Very large learning rate the step is too high made convergence speed very fast or unstable and the final SSE high or diverging

 * Very small learning rate made convergence speed Very slow and the final SSE Low but needs many iterations also if the learning rate is extremly small it may lead to fall in local minimum and made the SSE get high


```python
model_extremly_small_lr = LinearRegression_Hazem(alpha=0.0000001, num_iterations=200)
model_extremly_small_lr.fit(x, y)
model_extremly_small_lr.plot(x,y)
```

# What happens if the learning rate is too large?
* The model overshoots the minimum
* SSE may oscillate or increase
* The model may fail to converge (diverge)

# Modify the class to support multiple features

```python
import numpy as np
from Multip_linearReg import MultipLinearRegression_Hazem
```

```python
x = np.array([50, 60, 70, 80, 90])
y = np.array([150, 180, 210, 240, 270])

# Reshape for multiple regression class
X = x.reshape(-1, 1)

model = MultipLinearRegression_Hazem(alpha=0.0001, num_iterations=100)

model.fit(X, y)


print("Weights:", model.w)
print("Bias:", model.b)


predictions = model.predict(X)
print("Predictions:", predictions)

# Plot loss
model.plot_loss()
```
