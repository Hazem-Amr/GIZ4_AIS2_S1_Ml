```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1 , 2 , 3 , 4])
y = np.array([2 , 2.8 , 3.6 , 4.5])
```

```python
w = 0
b = 0
alpha = 0.01
num_iteration = 5
```

```python
from sklearn.linear_model import LinearRegression

MSE_vals = []
SSE_Values = []
for i in range (num_iteration):
    y_hat = w*x + b
    
    delta_w = 1/2 * np.sum((y_hat - y) * x) 
    delta_b = 1/2 * np.sum(y_hat - y)

    w = w - alpha * delta_w 
    b = b - alpha * delta_b

    SSE = (np.sum((y_hat - y)**2))
    SSE_Values.append(SSE)

    MSE = (1/len(x)) * (np.sum((y_hat - y)**2))
    MSE_vals.append(MSE)

    # if (i+1) % 20 == 0:
    #     print(f'iteration {i+1} , SSE: {SSE}')
print(SSE_Values)
print(MSE_vals)
```

```python
print(f'optimized parameters : bais = {b} , slope = {w}')
```

```python
plt.figure(figsize=(12 , 5))

plt.subplot(1 ,2 , 1)
plt.plot(range(num_iteration) , SSE_Values , label = 'SSE')
plt.xlabel('Iteraion')
plt.ylabel('SSE')
plt.title('SSE over Iteration')
plt.legend()

plt.subplot(1 ,2 , 2)
plt.scatter(x , y , color = 'blue' , label = "Data Points")
plt.plot(x , w*x + b , color = 'red' , label = 'Regression line')
plt.show()
```

```python
class LinearRegressionGD:

    def __init__(self, alpha=0.01, num_iterations=100):
        self.alpha = alpha
        self.num_iterations = num_iterations
        self.w = 0.0
        self.b = 0.0
        self.SSE_values = []

    def predict(self, x):
        return self.w * x + self.b

    def compute_gradients(self, x, y, y_hat):
        dw = np.mean((y_hat - y) * x)
        db = np.mean(y_hat - y)
        return dw, db

    def compute_sse(self, y, y_hat):
        return np.sum((y_hat - y) ** 2)

    def fit(self, x, y):
        for i in range(self.num_iterations):
            y_hat = self.predict(x)

            dw, db = self.compute_gradients(x, y, y_hat)

            self.w -= self.alpha * dw
            self.b -= self.alpha * db

            sse = self.compute_sse(y, y_hat)
            self.SSE_values.append(sse)

            if (i + 1) % 20 == 0:
                print(f"Iteration {i+1}, SSE: {sse}")

        return self


    def plot_training(self, x, y):
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        plt.plot(range(self.num_iterations), self.SSE_values, label="SSE")
        plt.xlabel("Iteration")
        plt.ylabel("SSE")
        plt.title("SSE over Iterations")
        plt.legend()


        plt.subplot(1, 2, 2)
        plt.scatter(x, y, color="blue", label="Data Points")
        plt.plot(x, self.predict(x), color="red", label="Regression Line")
        plt.legend()

        plt.show()

    def get_params(self):
        return {
            "slope": self.w,
            "bias": self.b
        }
```

```python
x = np.array([1 , 2 , 3 , 4])
y = np.array([2 , 2.2 , 3.6 , 4.7])

model = LinearRegressionGD(alpha=0.01, num_iterations=100)
model.fit(x, y)

print("Optimized parameters:", model.get_params())

model.plot_training(x, y)
```
