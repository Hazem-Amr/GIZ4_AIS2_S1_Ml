```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1 , 2 , 3 , 4])
y = np.array([2 , 2.2 , 3.6 , 4.7])
```

```python
w = 0
b = 0
alpha = 0.01
num_iteration = 20
```

```python
from sklearn.linear_model import LinearRegression

SSE_Values = []
for i in range (num_iteration):
    y_hat = w*x + b
    
    delta_w = 2 * np.sum((y_hat - y) * x) 
    delta_b = 2* np.sum(y_hat - y)

    w = w - alpha * delta_w 
    b = b - alpha * delta_b

    SSE = (np.sum((y_hat - y)**2))
    SSE_Values.append(SSE)

    if (i+1) % 20 == 0:
        print(f'iteration {i+1} , SSE: {SSE}')
```

```python
print(f'optimized parameters : bais = {b} , slope = {w}')
```

```python
plt.figure(figsize=(12 , 5))

plt.subplot(1 ,2 , 1)
plt.plot(range(num_iteration) , SSE_Values , label = 'SSE')
plt.xlabel('Iteraion')
plt.ylabel('SSE')
plt.title('SSE over Iteration')
plt.legend()

plt.subplot(1 ,2 , 2)
plt.scatter(x , y , color = 'blue' , label = "Data Points")
plt.plot(x , w*x + b , color = 'red' , label = 'Regression line')
plt.show()
```
