```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
x = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x,y)
y_pred_lin = lin_reg.predict(x)

y_pred_lin
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
ridge = Ridge(alpha=50.0)
ridge.fit(x,y)
y_pred_ridge = ridge.predict(x)

print(ridge.coef_)
print(ridge.intercept_)
```

```python
lasso = Lasso(alpha=50.0)
lasso.fit(x,y)
y_pred_lasso = lasso.predict(x)

print(lasso.coef_)
print(lasso.intercept_)
```

```python
print("Linear Regression:")
print("MSE:" , mean_squared_error(y , y_pred_lin))
print("R2:" , r2_score(y , y_pred_lin))
print("=============================================")
```

```python
print("ridge:")
print("MSE:" , mean_squared_error(y , y_pred_ridge))
print("R2:" , r2_score(y , y_pred_ridge))
print("=============================================")
```

```python
print("Lasso:")
print("MSE:" , mean_squared_error(y , y_pred_lasso))
print("R2:" , r2_score(y , y_pred_lasso))
print("=============================================")
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x,y , color='blue' , label = "Actual data")
plt.plot(x, y_pred_lin, color = "red" , label = "Linear Regression")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x,y , color='blue' , label = "Actual data")
plt.plot(x, y_pred_ridge, color = "red" , label = "Ridge Regression")
plt.title("Ridge Regression")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x,y , color='blue' , label = "Actual data")
plt.plot(x, y_pred_lasso, color = "red" , label = "Lasso Regression")
plt.title("Lasso Regression")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y , color = 'blue' , label = "actual")

plt.plot(x,y , y_pred_lin, color = 'red' , label = 'linear regression')
plt.plot(x,y , y_pred_ridge, color = 'green', linestyle = "--" , label = 'ridge regression')
plt.plot(x,y , y_pred_lasso, color = 'purple' ,linestyle = ":", label = 'lasso regression')

plt.title("Comparison of Linear , Rigde and Lasso Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
