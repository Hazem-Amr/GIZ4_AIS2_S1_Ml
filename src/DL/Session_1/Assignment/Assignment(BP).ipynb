```python
import numpy as np

# Activation function: Logistic (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative for backpropagation
def sigmoid_derivative(x):
    return x * (1 - x)

# 1. Initialize Inputs and Targets 
inputs = np.array([0.05, 0.10])
targets = np.array([0.01, 0.99])

# 2. Initialize Weights and Biases 
# W1: [w1, w2] for h1, [w3, w4] for h2
W_hidden = np.array([[0.15, 0.20], [0.25, 0.30]]) 
# W2: [w5, w6] for o1, [w7, w8] for o2
W_output = np.array([[0.40, 0.45], [0.50, 0.55]]) 

b1 = 0.35
b2 = 0.60
lr = 0.5 # Learning rate

def train(epochs):
    global W_hidden, W_output
    for i in range(epochs):
        # --- FORWARD PASS ---
        # Calculate hidden layer (h1, h2)
        net_h = np.dot(W_hidden, inputs) + b1
        out_h = sigmoid(net_h)
        
        # Calculate output layer (o1, o2)
        net_o = np.dot(W_output, out_h) + b2
        out_o = sigmoid(net_o)
        
        # Calculate Total Error [cite: 33]
        error = 0.5 * np.sum((targets - out_o)**2)
        
        if i == 0:
            print(f"--- Iteration 1 ---")
            print(f"Hidden Outputs: {out_h}")
            print(f"Network Outputs: {out_o}")
            print(f"Total Error: {error}\n")
        
        # --- BACKWARD PASS (Backpropagation) ---
        # Output Layer Gradients
        d_error_out_o = -(targets - out_o)
        d_out_o_net_o = sigmoid_derivative(out_o)
        delta_output = d_error_out_o * d_out_o_net_o
        
        # Hidden Layer Gradients
        delta_hidden = np.dot(W_output.T, delta_output) * sigmoid_derivative(out_h)
        
        # Update Weights
        W_output -= lr * np.outer(delta_output, out_h)
        W_hidden -= lr * np.outer(delta_hidden, inputs)

    return out_o

# Run training for 10,000 iterations 
final_pred = train(10000)
print(f"--- After 10,000 Iterations ---")
print(f"Final Prediction: {final_pred}")
```
