```python
# Multilayer Perceptron
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
```

```python
visible = Input(shape=(10,))
hidden1 = Dense(500, activation='tanh')(visible) #layer-1
hidden2 = Dense(300, activation='tanh')(hidden1)
hidden3 = Dense(100, activation='tanh')(hidden2)

output = Dense(1, activation='sigmoid')(hidden3)
model = Model(inputs=visible, outputs=output)
# summarize layers
print(model.summary())
```

# **Sigmoid function leads to Vanishing gradient**

In **back probagation** the sigmoid function gives me values between [0,0.25] this happen because when i go backward the sigmoid function drivative is (σ[1-σ])

and this leads to Vanishing gradinet so we will use another activation function Like (RELU) to prevent this problem or use optmization algorithms like adam optmizer

But RELU leads to Explosion but it's more fine than vanishing
