```python
import numpy as np
import pandas as pd

data = [
    [1, 444, 'abc'],
    [np.nan, 555, 'def'],
    [3, 666, 'ghi'],
    [4, np.nan, 'xyz']
]

df = pd.DataFrame(data, columns=['id', 'salary', 'dept'])
df
```
