```python
def array_reshape ( arr , rows = -1 , col = -1):

    try:
        arr = np.array(arr)
        return arr.reshape(rows , col)
    except (TypeError , ValueError) as e:
        raise ValueError(f'Invalid reshape operation:{e}')
    
```

```python
vec = [1,2,3,4,5,6]
print(array_reshape(vec , 2, 3))
```
