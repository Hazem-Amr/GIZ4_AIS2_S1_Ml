```python
class Passenger:
    def __init__(self, passenger_id, name):
        self.id = passenger_id
        self.name = name


class Flight:
    def __init__(self, number, destination, seats):
        self.number = number
        self.destination = destination
        self.seats = seats
        self.passengers = []


# flight1 = Flight(1, 'Cairo', 30)


    def book(self, passenger):
        if self.seats == 0:
            print('NO seats')
            return

        for p in self.passengers:
            if p.id == passenger.id:
                print("Already Booked")
                return
            
        self.passengers.append(passenger)
        self.seats -= 1
        print('Successfully Booked')


    def cancel(self, passenger_id):
        for p in self.passengers:
            if p.id == passenger_id:
                self.passengers.remove(p)
                self.seats += 1
                print("Booking cancelled")
                return
            
        print("Passenger Not Found")

    
    def show(self):
        print(f"the flight number={self.number}")
        print(f"the flight destination={self.destination}")
        print(f"the Available seats={self.seats}")

        if len(self.passengers) == 0:
            print('No Passengers')

        else:
            print('All Passengers: ')
            for p in self.passengers:
                print(p.id, '--', p.name)


flights = []

def find_flight(number):
    for f in flights:
        if f.number == number:
            f.show()
            return f
        
    return None


def add_flight():
    number = input('flight Number')
    destination = input('destination: ')
    seats = int(input('Seats'))
    Time = input('flight Time')

    new_flight = Flight(number, destination, Time, seats)
    flights.append(new_flight)
    print('Flight Added Successfully')

def Cancel_flight():
    number = input('flight number to cancel: ')
    f = find_flight(number)

    if f is None:
        print('flight Not founded')
        return
```
