```python
def calculate_Average():
    num_student = int(input('number of students'))
    grades = []

    for student in range (num_student):
        grade = float(input('enter the grade'))
        grades.append(grade)

    if num_student == 0 :
        print('no students')

    else:
        print(f'the average of grades = {sum(grades)/num_student}')

calculate_Average()
```

```python
def add(a,b):
    return a + b

def substract (a,b):
    return a - b

def mult(a,b):
    return a*b

def div(a,b):
    if b == 0:
        print('invalid division')

    else: 
        return a/b
    



while True:
    print('1- add / 2- sub / 3- mult / 4- div')
    choice = int(input('enter ur choice'))

    if choice == 0 :
        break

    a = float(input('enter the numer'))
    b = float(input('enter the number'))

    if choice == 1:
        print(add(a,b))

    elif choice == 2 :
        print(substract(a,b))

    elif choice == 3 :
        print(mult(a,b))

    elif choice == 4 :
        print(div(a,b))
```

```python
def count_char(text , char):
    char = char.lower()
    count = 0

    for x in text.lower():
        if x == char:
            count += 1

    return count

count_char("HeLlo python world" , "l")
```
