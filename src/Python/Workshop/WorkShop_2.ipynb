```python
def Tasks():

    Tasks = []

    while True :
        print(" \n 1.Add Task \n 2.Delete Task \n 3.Show Tasks \n 4.Exit \n")
        choice = input("Please enter your choice")

        if choice == "1":
            Task = input("enter your task")
            Tasks.append(Task)

        elif choice == "2":
            Task_delete_num = int(input('enter the task number you want to delete'))
            del Tasks[Task_delete_num - 1]

        elif choice == "3":
            for i , j in enumerate (Tasks , 1): 
                print(f"{i}: {j}")

        elif choice == "4":
            break

        else:
            print("Invalid Input")

Tasks()
```

```python
import random
def play_game_rps():
    choicess = ['paper' , 'rock' , 'scissor']
    score = {"player": 0 , "computer": 0 }

    for i in range(3):
        player = input(" 'paper' / 'rock' / 'scissor' ").lower().strip()
        computer = random.choice(choicess)
        print(f" the computer select {computer} the player select {player}")

        if player == computer:
            print("Draw\n")

        elif (player == "paper" and computer == "rock") or (player == "scissor" and computer == "paper") or (player == "rock" and computer == "scissor"):
            score["player"] += 1
            print ("player wins\n")

        else:
            score["computer"] += 1
            print("computer wins\n")

    print(f"\n the final score for player: {score["player"]} and computer score: {score["computer"]}")

play_game_rps()
```

```python
employees = {}

while True:
    name = input('Enter the name of employee (or exit)')
    if name.lower() == 'exit':
        break

    points = []
    hours = []

    for day in range(5):
        point = int(input(f'enter the points in day {day}'))
        hour = int(input(f'enter the hours in day {day}'))
        
        points.append(point)
        hours.append(hour)

        employees[name] = {
            'hours': hours ,
            'points': points ,
            'sum_points': sum(points),
            'sum_hours': sum(hours),
            'average points': sum(points)/5
                    }

for name , data in employees.items():
    print(f"{name}: sum points = {data['sum_points']} and sum hours = {data['sum_hours']}")

```
