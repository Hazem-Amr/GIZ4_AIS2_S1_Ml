```python
degree = float(input("Enter student degree (0-100): "))

while degree < 0 or degree > 100:
    degree = float(input("Invalid input Enter a value between 0 and 100: "))

if 90 <= degree <= 100:
    print("Grade: A")
elif 80 <= degree < 90:
    print("Grade: B")
elif 70 <= degree < 80:
    print("Grade: C")
elif 60 <= degree < 70:
    print("Grade: D")
else:
    print("Grade: F")
```
