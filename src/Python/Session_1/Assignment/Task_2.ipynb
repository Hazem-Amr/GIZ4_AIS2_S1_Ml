```python
email = "Amit_ml@gmail.edu" 

#Input Validation
validation = (email.count("@") == 1) and ("." in email.split("@")[1]) \
                                     and "valid" or "Invalid email" 
print(validation)

#Extract Username
User_name = email.split('@')[0]
print("user name:",User_name)

#Extract Domain
domain = (email.split("@")[1]).split(".")[0]
print("Domain:",domain)

#Check for Domain Ending
domain_type = email.endswith(".com") and "Commercial Domain" or \
              email.endswith(".edu") and "Educational Domain" or \
              "Other Domain"

print("Domain Type:",domain_type)
```

Task-2

```python
Encoded_Message = "###!!@emocleW EPGTQ!!!6789"

#Extract the core part of the message
core = (Encoded_Message.split('@')[1]).split("!")[0]
print("Core_message:",core)

#Reverse the first word
first_word = core.split()[0]
Rev_first_word = first_word[::-1]
print("Reversed first word:",Rev_first_word)

#Replace shifted vowels in the second word
second_word = core.split()[1]
decoded_second = second_word[1:]
print("Second word decoded:", decoded_second)

#Final decoded message
Final_decoded_message = Rev_first_word + " " + decoded_second
print("Final_decoded_message:", Final_decoded_message)
```

```python
Encoded_Message = "&&&**$gnirtS PLIO!!@1234"

#Extract the core part of the message
core = (Encoded_Message.split('$')[1]).split("!")[0]
print("Core_message:",core)

#Reverse the first word
first_word = core.split()[0]
Rev_first_word = first_word[::-1]
print("Reversed first word:",Rev_first_word)

#Replace shifted vowels in the second word
second_word = core.split()[1]
decoded_second = second_word.replace("I","E").replace("O","U")
print("Second word decoded:", decoded_second)

#Final decoded message
Final_decoded_message = Rev_first_word + " " + decoded_second
print("Final_decoded_message:", Final_decoded_message)
```
